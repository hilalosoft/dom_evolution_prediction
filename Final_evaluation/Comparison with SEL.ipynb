# Counting how many timestimes the locator survived

#### to identify how many false cases can be skipped in a consecutive version change the variable called fail_margin

```python
import pandas as pd
import os

project_comparison={}
for project in os.listdir("C:\\Users\\hilal.taha\\PycharmProjects\\pythonProject\\database\\statistics\\evaluation_data\\Final_evaluation\\locators_data\\"):
    print("---------------"+project+ "-----------------")
    df = pd.read_csv("C:\\Users\\hilal.taha\\PycharmProjects\\pythonProject\\database\\statistics\\evaluation_data\\Final_evaluation\\locators_data\\"+project, delimiter=",")
    fail_margin = 1
    elements_comparison=[]
    reloc_dictionary={}
    failed_dictionary={}
    for row in df.iterrows():
        RELOC_score=0
        for column in df.columns:
            reloc_dictionary[column]=0
            failed_dictionary[column]=0
    for index, row in df.iterrows():
        RELOC_score=0
    #     print(row)
        for column in df.columns:
            value = df.iloc[row[0],df.columns.get_loc(column)]
            # Access the count of True values
            if value:
                if failed_dictionary[column]<=fail_margin:
                    reloc_dictionary[column]+=value
                    failed_dictionary[column]=0
            else:
                failed_dictionary[column]+=1
    RELOCvSel=0
    highest_Sel=0
    for key in reloc_dictionary:
        if(key[-5:])=="RELOC":
            reloc_score=RELOCvSel-highest_Sel
            print("relocvSEL: "+ str(reloc_score))
            elements_comparison.append(reloc_score)
            RELOCvSel=reloc_dictionary[key]
            highest_Sel=0

        if(key[-3:])=="SEL":
    #         print("SEL: "+str(reloc_dictionary[key]))
            if reloc_dictionary[key]>highest_Sel:
                highest_Sel=reloc_dictionary[key]
    row_count = df.shape[0]
    final_score=0
    winning_survivability=0
    for element in  elements_comparison:
        if element >0:
            winning_survivability+=1
        elif element <0:
            winning_survivability+= -1
        final_score+= element
    elements_comparison.append(final_score)    
    elements_comparison.append(winning_survivability)
    elements_comparison.append(row_count-1)
    project_comparison[project]= elements_comparison
```

# Saving the info inside a file
So I am saving for each project a row, each row is 20 elements. The score of RELOC on each element is how long the locator survived RELOC- SELENIUM, as such, if the value is >0 then RELOC survived longer, if <0 then it means Selenium survived longer.
The last two Columns are, total survavability of all elements for a row( per project) then  the number of timestamps that were tested respectively

```python
import csv
filename="RELOCvSEL.csv"
with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Project'] + list(range(1, len(max(project_comparison.values(), key=len))+1)))  # Write header row
    for i, key in enumerate(project_comparison, start=1):
        row_values = [key] + project_comparison[key]
        writer.writerow(row_values)
```
